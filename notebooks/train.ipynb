```python
IMAGE_DIR = '../data/original_images/'
ANNOTATIONS = '../data/annotations.parquet'
```

```python
import sys
sys.path.append('../src/')

import cv2 as cv
import numpy as np
from datasets import CRCNNEDataset
from skimage.segmentation import mark_boundaries
from matplotlib import pyplot as plt
```

```python
dataset = CRCNNEDataset(IMAGE_DIR, ANNOTATIONS)
image, labels, image_path = dataset[0]
print('dataset (length):', len(dataset))
print('image (path) (size) (type):', image_path, image.size, type(image))
print('labels (shape):', labels.shape)
```

```python
new_image = cv.rectangle(cv.imread(image_path), (1073, 971), (47 + 1073, 39 + 971), (0, 255, 0), 2)
plt.imshow(cv.cvtColor(new_image, cv.COLOR_BGR2RGB))
plt.axis('off')
```

```python
new_image = cv.imread(image_path)
for label in labels:
    new_image = cv.rectangle(new_image, (label[2], label[3]), (label[1] + label[2], label[0] + label[3]), (0, 255, 0), 2)
```
