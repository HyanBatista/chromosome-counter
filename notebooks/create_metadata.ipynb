```python
METADATA = '../data/informations_about_the_cutouts.json'
```

```python
import cv2 as cv
import json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_json(METADATA)
df.head()
```

```python
data = {}

for _, row in df.iterrows():
    if not row['image'] in data:
        data[row['image']] = []
    data[row['image']].append([row['cord_h'], row['cord_w'], row['cord_x'], row['cord_y']])
```

```python
new_data = []
for key, value in data.items():
    new_data.append({
        'image': key,
        'labels': value
    })
chromosomes = pd.DataFrame(data=new_data)
chromosomes.head()
```

```python
chromosomes.to_parquet('../data/annotations.parquet')
```
